```python
# CounterVectorizer에 넣기 위해서 문자열로 변환

movies_df['genres_literal'] = movies_df['genres'].apply(lambda x : (' ').join(x))
```

```python
movies_df.head()
```

```python
# Countervectorizer : 문자열 => vector => sparse matrix

#          tok1,  tok2, ...,  tok50000

# doc1      1 0 0 0 0 1 ... 2 ...
# doc2      0 2 0 0 0 0 0 0 0 1 0 ..
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vect = CountVectorizer(min_df = 0, ngram_range=(1, 2))
genre_mat=vect.fit_transform(movies_df['genres_literal'])
```

```python
genre_mat.toarray().shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
genre_sim = cosine_similarity(genre_mat, genre_mat)
```

```python
genre_sim
```

```python
genre_sim_sorted_ind = genre_sim.argsort()[:, ::-1]
genre_sim_sorted_ind
```

```python
movie_index = 50
movies_df.loc[genre_sim_sorted_ind[movie_index][:5]]
```

```python
def find_sim_movie(df, sorted_ind, title_name, top_n = 30):
    
    title_movie = df[df['title'] == title_name]          # 해당 영화데이터 row
    title_index = title_movie.index.values               # 해당 영화의 index
    
    similarity_indexex = sorted_ind[title_index, :(top_n)]   
    # sorted_ind matrix : row - 영화인덱스 순서, columns - 유사도 순으로 올림차순 영화인덱스
    # sorted_ind : 유사도 matrix[해당영화 row index, 왼쪽에서 n개 칼럼]    
    
    similarity_indexexs = similarity_indexex.reshape(-1)   # 유사도가 높은 영화인덱스들
    
    return df.iloc[similarity_indexexs]      # 유사도 높은 영화들의 데이터를 리턴
```

```python
df = movies_df
sorted_ind = genre_sim_sorted_ind
title_name = 'The Godfather'
top_n = 5

find_sim_movie(df, sorted_ind, title_name, top_n)
```

```python
def weighted_vote_average(record):
    v = record['vote_count']
    r = record['vote_average']
    
    record = record[(v >record['vote_count'].mean()) &  (r > record['vote_average'].mean())]
    
    record = record.sort_values('popularity',ascending=False) 
        
    return record
```

```python
def find_sim_movie_with_popularity(df, sorted_ind, title_name, top_n = 10):
    
    title_movie = df[df['title'] == title_name]          # 해당 영화데이터 row
    title_index = title_movie.index.values               # 해당 영화의 index
    
    similarity_indexex = sorted_ind[title_index, :(top_n) * 3]   
    # sorted_ind matrix : row - 영화인덱스 순서, columns - 유사도 순으로 올림차순 영화인덱스
    # sorted_ind : 유사도 matrix[해당영화 row index, 왼쪽에서 n개 칼럼]    
    
    similarity_indexexs = similarity_indexex.reshape(-1)   # 유사도가 높은 영화인덱스들
    
    record =  df.iloc[similarity_indexexs]      # 유사도 높은 영화들의 데이터를 리턴
    
    final_df = weighted_vote_average(record)
    
    return final_df
```

```python
df = movies_df
sorted_ind = genre_sim_sorted_ind
title_name = 'The Godfather'
top_n = 5


find_sim_movie_with_popularity(df, sorted_ind, title_name, top_n = 10)
```
